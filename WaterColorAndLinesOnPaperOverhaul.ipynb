```python
#%%py5drawsvg -f test.svg 400 400
import numpy as np

p = 100
mv = 50
colmv = 2000
alph = 1
global h1, h, s1, s, b1, b, i, j, h2

rez = .00087 

scaling = 80
resolution = .0026
npoints = 500

radius = 80
nrings = 15
bcol = 0

def setup():
    
    size(1024,1024)
    color_mode(HSB, 45, 50, 100, 255)
    no_stroke()
    maxVary = 5
    i, j, h, s, b, h1, s1, b1, hMax, hMin, sMax, sMin, bMax, bMin = get_color(maxVary)
    numb = (width*height) / 2500
    
    background(h1, s1-30,b1+30)
    fill_background(numb, i, j, h, s, b, hMax, hMin, sMax, sMin, bMax, bMin)
    i, j, h, s, b, h1, s1, b1, hMax, hMin, sMax, sMin, bMax, bMin = get_color(maxVary)
    numb = (width * height) / 250
    fill_background(numb, i, j, h, s, b, hMax, hMin, sMax, sMin, bMax, bMin)
    
    apply_filter(BLUR, 0.5)
    paper_texture()
    
    rows = 4
    columns = 4
    gridres = (rows*columns)^2
    for k in range(columns):
        for p in range(rows):
            if k == p or k + p % 2 == 0 :
                #pass
                center = k * 512 + 24
                center2 = p * 512 + 256
                draw_tree(center, center2, nrings, scaling, radius)
            else:
                center = k * 512 + 256
                center2 = p * 512 + 256
                draw_tree(center, center2, nrings, scaling, radius)
    

    #draw_varying_width_line(40, 40, 900, 900, 4, 20)
    

def draw_tree(x1, y1, nrings, scaling, radius):
    stroke_weight(2)
    for r in np.arange(0, radius, radius/nrings * cos(random())):    
        begin_shape()
        for a in np.arange(-TAU/npoints, TAU+ TAU/npoints + .01, TAU/npoints):
            x = x1 *.5 + r * cos(a) 
            y = y1 *.5 + r * sin(a)
        
            rnoise = np.interp([noise(x * resolution), noise(y * resolution)], [0,1], [-scaling, scaling])
            #point(x  + rnoise[0],y + rnoise[1])
            stroke(bcol* cos(random(0,1)), bcol* cos(random(0,1)), bcol* cos(random(0,1))) #- 30
            curve_vertex(x  + rnoise[0],y + rnoise[1])
    
            if(random() > 0.75-0.25*sin(r)):
                end_shape()
                begin_shape()
            
            #if(random() > 0.75-0.25*cos(r)):
            #    end_shape()
            #    begin_shape()
            
        end_shape()
        
    no_loop()

def draw_varying_width_line(x1, y1, x2, y2,  start_width, end_width):
    fill(400)
    half_start_width = start_width / 2
    half_end_width = end_width / 2
    begin_shape()
    curve_vertex(x1 + half_start_width, y1 - half_start_width)
    vertex(x2 + half_end_width, y2 - half_end_width)
    vertex(x2 - half_end_width, y2 + half_end_width)
    vertex(x1 - half_start_width, y1 + half_start_width)
    vertex(x1 + half_start_width, y1 - half_start_width)
    #curve_point(x1, y1, x2, y2, end_width)
    end_shape()

    
def get_color(maxVary):
    h = random(30,60)
    h1 = h
    hMax = h1 + maxVary
    hMin = h1 - maxVary
    s = 21
    s1 = s
    sMax = 25
    sMin = 18
    b = 70
    b1 = b
    bMax = 100
    bMin = 70
    i = random(width)
    j = random(height)
    
    return i, j, h, s, b, h1, s1, b1, hMax, hMin, sMax, sMin, bMax, bMin

def fill_background(numb, i, j, h, s, b, hMax, hMin, sMax, sMin, bMax, bMin):
    for q in np.arange(0, numb):
        i = i + random(-mv, mv)
        j = j + random(-mv, mv)
        h = h + random(-colmv, colmv)
        s = s + random(-colmv*2, colmv*2)
        b = b + random(-colmv*2, colmv*2)
        
        if i < 0:
            i = width
        elif i > width:
            i = 0
        if j < 0:
            j = height
        elif j > height:
            j = 0
        if h > hMax:
            h = hMax
        elif h < hMin:
            h = hMin
        h2 = h
        if h > 90:
            h2 = h - 90
        elif h < 0:
            h2 = h + 22
        if s > sMax:
            s = sMax
        if s < sMin:
            s = sMin
        if b > bMax:
            b = bMax
        elif b < bMin:
            b = bMin
        
        fill(h2, s, b, alph)
        push()
        translate(i, j)
        rotate(random(PI*2))
        begin_shape()
        for m in np.arange(0, PI*2, 1):
            r = random(20,50)
            x = cos(m) * r
            y = sin(m) * r
            vertex(x, y)
            
        end_shape(CLOSE)
        pop()

def paper_texture():
    no_fill()
    texture_num = width*height/50
    
    for i in np.arange(0, texture_num):
        stroke(random(15,75),random(0,50),random(100,150),3)
        x = random(-width*0.2,width*1.2)
        y = random(-height*0.2,height*1.2)
        push()
        translate(x,y)
        stroke_weight(2)
        point(0,0)
        stroke_weight(1)
        rotate(random(PI*2))
        
        curve(random(60,220),0,0,random(-50,50),random(-50,50),random(60,120),random(60,120),random(60,220))
        
        pop()
    

run_sketch()
```
