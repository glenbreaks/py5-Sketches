```python
import numpy as np

p = 1
mv = 50
colmv = 2
alph = 5
global h1, h, s1, s, b1, b, i, j, h2

rez = .00087 

scaling = 100
resolution = .002
npoints = 100

radius = 400
nrings = 10

rez = .00087 

scaling = 100
resolution = .002
npoints = 100

radius = 400
nrings = 20
bcol = 150

def setup():
    
    size(1024,1024)
    color_mode(HSB, 360, 120, 100, 255)
    no_stroke()
    maxVary = 20
    i, j, h, s, b, h1, s1, b1, hMax, hMin, sMax, sMin, bMax, bMin = get_color(maxVary)
    numb = (width*height) / 25
    
    background(h1, s1-10,b1-30)
    fill_background(numb, i, j, h, s, b, hMax, hMin, sMax, sMin, bMax, bMin)
    i, j, h, s, b, h1, s1, b1, hMax, hMin, sMax, sMin, bMax, bMin = get_color(maxVary)
    numb = (width * height) / 250
    fill_background(numb, i, j, h, s, b, hMax, hMin, sMax, sMin, bMax, bMin)
    
    apply_filter(BLUR, 0.5)
    paper_texture()
    
    for r in np.arange(0, radius, radius/nrings * cos(random())):    
        begin_shape()
        for a in np.arange(-TAU/npoints, TAU+ TAU/npoints + 1, TAU/npoints):
            x = height *.5 + r * cos(a) 
            y = width *.5 + r * sin(a)
        
            rnoise = np.interp([noise(x * resolution), noise(y * resolution)], [0,1], [-scaling, scaling])
            #point(x  + rnoise[0],y + rnoise[1])
            stroke(bcol* cos(random(0,.55)), bcol* cos(random(0,.55)), bcol* cos(random(0,.55)))
            curve_vertex(x  + rnoise[0],y + rnoise[1])
    
            if(random() > 0.75-0.25*sin(r)):
                end_shape()
                begin_shape()
            
            #if(random() > 0.75-0.25*cos(r)):
            #    end_shape()
            #    begin_shape()
            
        end_shape()
        
    no_loop()
    
def get_color(maxVary):
    h = random(360)
    h1 = h
    hMax = h1 + maxVary
    hMin = h1 - maxVary
    s = 90
    s1 = s
    sMax = 100
    sMin = 40
    b = 70
    b1 = b
    bMax = 70
    bMin = 30
    i = random(width)
    j = random(height)
    
    return i, j, h, s, b, h1, s1, b1, hMax, hMin, sMax, sMin, bMax, bMin

def fill_background(numb, i, j, h, s, b, hMax, hMin, sMax, sMin, bMax, bMin):
    for q in np.arange(0, numb):
        i = i + random(-mv, mv)
        j = j + random(-mv, mv)
        h = h + random(-colmv, colmv)
        s = s + random(-colmv*2, colmv*2)
        b = b + random(-colmv*2, colmv*2)
        
        if i < 0:
            i = width
        elif i > width:
            i = 0
        if j < 0:
            j = height
        elif j > height:
            j = 0
        if h > hMax:
            h = hMax
        elif h < hMin:
            h = hMin
        h2 = h
        if h > 360:
            h2 = h - 360
        elif h < 0:
            h2 = h + 360
        if s > sMax:
            s = sMax
        if s < sMin:
            s = sMin
        if b > bMax:
            b = bMax
        elif b < bMin:
            b = bMin
        
        fill(h2, s, b, alph)
        push()
        translate(i, j)
        rotate(random(PI*2))
        begin_shape()
        for m in np.arange(0, PI*2, 1):
            r = random(20,50)
            x = cos(m) * r
            y = sin(m) * r
            vertex(x, y)
            
        end_shape(CLOSE)
        pop()

def paper_texture():
    no_fill()
    texture_num = width*height/50
    
    for i in np.arange(0, texture_num):
        stroke(random(100,150),random(100,150),random(100,150),6)
        x = random(-width*0.2,width*1.2)
        y = random(-height*0.2,height*1.2)
        push()
        translate(x,y)
        stroke_weight(3)
        point(0,0)
        stroke_weight(1)
        rotate(random(PI*2))
        
        curve(random(60,220),0,0,random(-50,50),random(-50,50),random(60,120),random(60,120),random(60,220))
        
        pop()
    

run_sketch()
```
